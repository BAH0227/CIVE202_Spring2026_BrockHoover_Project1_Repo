## Step 3:
Next, we then find the effect of temperature, humidity, and altitude on air quality, functions are used on the air data to accomplish this task.


**Temperature Categorization:**

Temperature was categorized according to project instructions, with a temperature below 32 being designated as freezing, a temperature between 32 and 50 being designated as cool, and a temperature between 51 and 70 being designated as hot, and any temperature higher than 70 being designated as hot

```python
def cat_temperature(temperature):
    if temperature < 32:
        return 'Below Freezing'
    elif 32 <= temperature <= 50:
        return 'Cool'
    elif 51 <= temperature <= 70:
        return 'Warm'
    else:
        return 'Hot'
air_data['cat_temperature'] = air_data['temperature'].apply(cat_temperature)

air_dataS
```

**Humidity Categorization:**

Humidity was categorized by the requested project instructions, with a humidity being considered low if the relative humidity was below 50, high if between 50 and 80, and very high if above 80

```python
def cat_humidity(humidity):
    if humidity < 50:
        return 'Low'
    elif 50 <= humidity <= 80:
        return 'High'
    else:
        return 'Very High'
air_data['cat_humidity'] = air_data['humidity'].apply(cat_humidity)

air_data
```

**Altitude Categorization:**

Nebraska doesn't have any extremely low or high altitudes, and the project requests don't ask for specific bounds. The lowest point in Nebraska is at 840 feet above sea level, and the highest point is 5,242 feet above sea level (which was found from https://nebraskalegislature.gov/pdf/bluebook/25-37.pdf). Setting ranges of 800 to 2300 feet, 2300 feet to 3800 feet, and 3800 to 5300 feet should allow us to get a good idea of how relative altitude in Nebraska could affect air quality

```python
def cat_altitude(sensoraltitude):
    if 800 <= sensoraltitude < 2300:
        return 'Relative Low'
    elif 2300 <= sensoraltitude <= 3800:
        return 'Relative Medium'
    else:
        return 'Relative High'
air_data['cat_altitude'] = air_data['sensor.altitude'].apply(cat_altitude)

air_data
```
